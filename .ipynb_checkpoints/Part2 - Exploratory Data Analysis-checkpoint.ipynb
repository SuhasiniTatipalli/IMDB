```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
# Open final_tmdb_data_2000 saved file and preview again
df_2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz", low_memory = False)
df_2000.head()
```

```python
# Open final_tmdb_data_2001 saved file and preview again
df_2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz", low_memory = False)
df_2001.head()
```

```python
#Concatenate the datafrme 2000 and 2001 into 1 dataframe

df_concat = pd.concat([df_2000, df_2001], axis=0)
```

```python
df_concat.head()
```

```python
print(len(df_2000))
print(len(df_2001))
```

```python
len(df_concat)
```

```python
#How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
budget_f= df_concat[(df_concat['budget']>0)]
budget_f.head()
len(budget_f)
```

```python
revenue_f= df_concat[(df_concat['revenue']>0)]
revenue_f.head()
len(revenue_f)
```

```python
#Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.

budget_Rev= df_concat[(df_concat['budget']>0) & (df_concat['revenue']>0)]
```

```python
budget_Rev
```

```python
#How many movies are there in each of the certification categories (G/PG/PG-13/R)?
budget_Rev['certification'].value_counts(dropna=False)
```

```python
#What is the average revenue per certification category?
budget_Rev.groupby(['certification'])['revenue'].mean()
```

```python
#What is the average budget per certification category?
budget_Rev.groupby(['certification'])['budget'].mean()
```
